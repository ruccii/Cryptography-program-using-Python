# Developing a Cryptography program in python

## Some important terminology:

- **Cryptography:** Changing the plane message into something anyone can not understand except the person who is having a access key to crack it. Also it can be defined as the art of writing or solving codes.
- **Cipher Text:** A text which is the result of an Encryption.
- **Encryption:** Converting a normal message to a secret message.
- **Decryption:** Converting back the secret message to a normal message.

## Problem statement:

- To write a python program that converts any given text to a ciphertext and when the key is available, it should convert the ciphertext back to the normal message.

## Steps to solve:

- Install a Library which can convert a Normal text to Ciphertext.
- Take Input from the User and convert it to Ciphertext.
- Display back the Ciphertext to the user.
- Load the Key and if the Key is same, based upon the Input provided, convert it to a normal Text.
